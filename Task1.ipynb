### Problem Definition 
#### Clean and Prepare raw Data for analysis

```python
import pandas as pd 
```

```python
df = pd.read_csv("Titanic_Dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df=df.drop_duplicates()
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df.rename(columns={
    'PassengerId': 'Passenger_ID',
    'Pclass': 'Passenger_Class',
    'Sex': 'Gender'
}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("Cleaned_Titanic_Dataset.csv",index=False)
```

### Key Insights 
#### The Age column contained multiple missing values.
#### The Cabin column had a very high percentage of null entries.
#### Duplicate records were identified and removed.
#### Data types were optimized for better analysis.
#### The overall quality and consistency of the dataset improved significantly after preprocessing."
